```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import dataframe_image as dfi
```

```python
covid_data = pd.read_csv('input/covid19_fullgrouped.csv')
covid_data["Date"] = pd.to_datetime(covid_data['Date'])
covid_data['YearMonth'] = covid_data['Date'].map(lambda date: 100*date.year + date.month)
covid_data

```

```python
covid_d = covid_data.query('YearMonth == 202004')
covid_d
```

```python
covid_c = covid_d.groupby('Country')['Active'].max().reset_index()
covid_c_sorted = covid_c.sort_values('Active',ascending=False).reset_index(drop=True)
covid_most_patient = covid_c_sorted.head(9)
dfi.export(covid_most_patient,'output/most_covid19_patients.png',table_conversion="matplotlib")
```

#### Death Rate in Iran

```python

# covid_data['Country'] = covid_data.query()
covid_d4_ir = covid_data.query('YearMonth == 202004 and Country=="Iran"').reset_index(drop=True)

fig, ax = plt.subplots(figsize=(12,6))

ax.stackplot(covid_d4_ir["Date"],covid_d4_ir["New cases"].to_numpy(),color='skyblue')
ax.stackplot(covid_d4_ir["Date"],covid_d4_ir["New deaths"].to_numpy(),color='darkorange')


plt.title('Death Rate in Iran per day')
fig.tight_layout()
fig.savefig('output/iran_death_rate.png')
covid_d4_ir
```
