```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
N = 500
std = 0.5

balloon = pd.DataFrame({
    'length': np.random.normal(35,std,N),
    'width': np.random.normal(30,std,N),
    'weight': np.random.normal(0.01,std-.2,N),
})
melon = pd.DataFrame({
    'length': np.random.normal(35,std,N),
    'width': np.random.normal(30,std,N),
    'weight': np.random.normal(2,std,N),
})

all = pd.concat([balloon,melon])
```

```python

fig = plt.figure()

ax = fig.add_subplot(projection='3d')
ax.scatter(balloon['length'], balloon['width'], balloon['weight'], label='balloon',alpha=.5,c='hotpink',marker='v')
ax.scatter(melon['length'], melon['width'], melon['weight'], label='melon',alpha=.5,c='g')
# ax.errorbar(balloon['length'], balloon['width'], balloon['weight'],.2, 0.2)
ax.set_xlabel('length(cm)')
ax.set_ylabel('width(cm)')
ax.set_zlabel('weight(kg)')
ax.legend()
plt.savefig('output/balloon_malon.png')
plt.show()
```
