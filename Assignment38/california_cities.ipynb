```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
cities = pd.read_csv('california_cities.csv')
```

```python
latitude, longitude = cities["latd"], cities["longd"]
population, area = cities["population_total"], cities["area_total_km2"]

seaborn.set()
plt.scatter(latitude,longitude,
            s=area,
            alpha=.5,linewidths=0
            ,c=np.log10(population)
            ,cmap='viridis')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Area and Population of Californian cities')
plt.grid()
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3, 7)

for area in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=area, label=str(area) + 'km$^2$')


plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Areas')
plt.grid()
plt.savefig('output/california_cities.png')
plt.show()
```
